```python
import argparse
import json

import requests
import pandas as pd


URL = "https://www.iaa.gov.il/umbraco/surface/FlightBoardSurface/Search"

session = requests.Session()

headers = {
    "User-Agent": "Mozilla/5.0",
    "X-Requested-With": "XMLHttpRequest",
    "Origin": "https://www.iaa.gov.il",
    "Referer": "https://www.iaa.gov.il/airports/ben-gurion/flight-board",
    "Content-Type": "application/x-www-form-urlencoded; charset=UTF-8",
}

payload = {
    "g-recaptcha-response": "",
    "FlightType": "Incoming",   # Change to Outgoing if needed
    "AirportId": "LLBG",
    "UICulture": "he-IL",
    "City": "",
    "Country": "",
    "AirlineCompany": "",
    "FromDate": "2/3/2026",  # You can set date here like "01/03/2026"
    "ToDate": "7/3/2026",
    "ufprt": ""  # Leave empty for first test
}

response = session.post(URL, headers=headers, data=payload)

print("Status code:", response.status_code)
with open("response.json", "w", encoding="utf-8") as f:
    f.write(response.text)
data = json.loads(response.text)
```
