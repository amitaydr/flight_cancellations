```python
gov_api_url = "https://data.gov.il/api/3/action/datastore_search?resource_id=e83f763b-b7d7-479e-b172-ae981ddc6de5"
session = requests.Session()

response = session.get(gov_api_url)
print("Status code:", response.status_code)
with open("data_gov.json", "w", encoding="utf-8") as f:
    f.write(response.text)
data_gov = json.loads(response.text)
```

```python
with open("data_gov.json", 'r') as f:
    data_gov = json.loads(f.read())
data_gov.keys()
```

```python
len(data_gov['result']['records'])
```

```python
data_df = pd.DataFrame(data_gov['result']['records'])
incoming = data_df[data_df["CHAORD"] == "A"]
outgoing = data_df[data_df["CHAORD"] == "D"]
```

```python
result = incoming.groupby("CHLOCCT").agg({"_id": "count"}).sort_values("_id", ascending=False)
result.columns = ["count"]
result["percent"] = (result["count"] / result["count"].sum() * 100).round(2)
result.reset_index().rename(columns={"CHLOCCT": "Country"})
```

```python
result.to_csv("incoming_flights_by_country.csv", index=True)
```
